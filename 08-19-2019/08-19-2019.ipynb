# [Supervised learning][1]
    

We have [labeled data][2] about the subject (i.e., patient or visit). That is, we know some truth values and we want to learn how to predict this truth value.  

### Approach:
1. It is a task of inferring a function from labeled training data.
2. Each data point is an example or a vector of features or values of attributes/covariates (predictors) and label/response (outcome) variable.
3. The function maps **<u>the training examples</u>** to **<u>the new unseen or testing examples</u>** to a specific label. 


### Notations:

1. Maping function to the training data is called **<u>model fitting</u>**.
2. Testing the model on unseen examples is called **<u>model generalization</u>**.
3. We need to minimize fitting and generalization errors. The typical curves of training and testing is shown [below][3]
4. **<u>Model paramters</u>** are the values or weights that are estimated from the data; but, **<u>model hyperparameters</u>** are the variables of the learners that could not be inferred from the data-- [detailed explanation][4]. 
5. Model complexity is the number of model parameters or degree of freedom; for example, quadratic model is more complex than linear one.

### <span style="color:red">Challenges:</span> 

1. Enough labeled data.
2. External data to validate. 
3. Data quality like Noise/outliers, non-linearity.

# [Unsupervised learning][1]


1. We **DO NOT** have labeled data about the subject. The aim is to understand the data identifying or visualizing similar groups, hidden patterns, or important features/attributes.
2. Many data scientists prefer to start solving AI problems by conducting unsupervised learning on the data to explore hidden relationships.

### Approach:

1. It is a descriptive task of inferring  similar groups or hidden patterns from the data.
2. Each data point is an example or a vector of features or values of attributes/covariates.
3. If label/response variable exists, then it is used only in the evaluation not in the learning.
4. The similarity function maps the examples to similar groups or hidden patterns, or important features/attributes.

### Approach types:

1. Discovering subpopulations (clustering).
2. Using visualization to explore the data (visual analytics).
3. Identifying important correlations among features or principle features (dimensionality reduction).
4. Discovering Frequent association and patterns among features (association/pattern mining).

# [Semi-supervised learning][1]



We have **a little bit of labeled data and a lot of unlabeled data** about the subject. The aim is use the labeled data to boost the unlabeled data  to do a specific learning task.

### [Approach:][1]

### [Approach 1.1]
1. Develop an Integration task between unsupervised and supervisor learners.
2. There are many versions with different orders of learning algorithms.  The aim is to boost the whole learning. 

### [Approach 1.2]
1. Use labeled seeds as initial examples for clustering. These seeds should be sampled using the positive and negative ratios from the data.
2. Run a clustering algorithm to find out the clusters around the seeds. 
3. Repeat the above two steps to refine and redistribute the clusters.

# [Learning strategies][1]

After we select the most adequate machine learning type that will solve our research question and we prepare our data set, we may think to choose one of the each following two strategies: 

1. Batch versus online (incremental) learning.

2. Instance-based (lazy) versus  model-based (eager) learning.

### Learning Strategy <span style="color:blue">:Batch Learning</span> 

The holistic data is provided to the learning algorithm at once.

#### Advantages:
1. Many algorithms need a lot of data to learn accurately.
2. Solutions over the whole data are usually simple.

#### Disadvantages:
1. Learning or training is effected by the whole data rare, incomplete, noises, or outlier examples.
2. These problems are usually happened in medical data sets.
3. Learning or training is time-consuming.
4. Learning or training requires a lot of computing resources such memory/disk spaces or CPU cycles.


### Learning Strategy <span style="color:blue">:Online (Incremental) Learning</span> 

Mini-batch(s)/example(s) is provided to the learning algorithm at a time (incrementally).

#### Advantages:
1. Updating learning process becomes easier.
2. Learn from the errors if any or do nothing.
3. Tackling the disadvantages of batch learning.

#### Disadvantages:
1. Many algorithms need a lot of data to learn accurately.

# [Main ML challenges][1]

Besides each machine learning task type challenges, we have the following common challenges: 
1. Insufficient quantitative training data:
    - Many algorithms need a lot of data to learn accurately.
    - In practice, the data size is small or medium. 
    - Some recommendations to develop synthesis data but it is too difficult.

<img style="float: center" src="./images/Size-accuracy.png" alt="drawing" Hight="200" width="300"/>

2. Insufficient qualitative training data
    - The lack of representative examples for each label.
    - The sampling of examples is biased such that it excludes important population characteristics.
    - Some examples have missing values and/or they are rare/noise/outlier. 
        - Do we need to handle these cases? How?
        - Do we need to apply methods to preprocess the data? Usually, yes!
    - Irrelevant features/attributes.
        - Do we need to remove or combine some of them? 
    - Data could have some non-linearity or fluctuation trends. 
        - How to identify these trends?
3. Model complexity, assumptions, and parameters: If we select model-based learning, then 
    - We need to care of the following theorem:
        - __No Free Lunch (NFL) theorem:__ If we make absolutely no assumption about the data, then there is no reason to prefer one model over any other.  There is no model that is a priori guaranteed to work better. There is always a need to select the best model for each data set; if possible. Otherwise, we may  try to generalize the model.
    - We need to:
        - Identify potential data assumptions that could handle the mentioned deficiencies in data quantity and quality. 
        - Select the most adequate learning algorithm to handle these assumptions:
            - Need to optimize the algorithm parameters (i.e., hyper-parameters) settings.
            - Develop the best performing (i.e., minimizing the loss function or maximizing the accuracy) model with its parameters’ values.
    - We need to select between simple and complex model in order to minimize:
        - __Under-fitting:__  In the training data, the model is too simple to discover important hidden relationships among features . So, its  performance is low.
        - __Over-fitting:__  In the test data, the model developed in the training data could not generalize what it learned to recognize new cases correctly. Indeed, the model may be too complex such that it memorized <u> every training case relationship and data noise </u> rather than it drew some generalizations from the training data.  The model performance is high in the training data but it is low in test or unseen data since it generates a lot of generalization errors.

```python
#Visualize Pearson pairwise correlation of features
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(corr_matrix,cmap='coolwarm', vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = np.arange(0,len(iris.drop(["Id","Species"], axis=1).columns),1)
ax.set_xticks(ticks)
plt.xticks(rotation=90)
ax.set_yticks(ticks)
ax.set_xticklabels(iris.drop(["Id","Species"], axis=1).columns)
ax.set_yticklabels(iris.drop(["Id","Species"], axis=1).columns)
plt.show()
```
